```python
from auto_encoder import AutoEncoder

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_path = "/content/drive/MyDrive/VQ-VAE-FashionMNIST/results/unshuffled_MNIST/model.pth"
model = AutoEncoder(device, configuration)
model.load_state_dict(torch.load(model_path, map_location=device), strict = False)
```

```python
decoder_model = model._decoder
vq_vae_model = model._vq_vae
```

```python
from matplotlib import pyplot as plt

decoder_model.eval()
vq_vae_model.eval()
decoder_model.to("cpu")
vq_vae_model.to("cpu")

num_embeddings = vq_vae_model._num_embeddings
embedding_dim = vq_vae_model._embedding_dim

print(embedding_dim)

quant_vectors = vq_vae_model._embedding.weight.data

indices = torch.randint(0, num_embeddings, (64,64))
tirage_prior = quant_vectors[indices].unsqueeze(0).permute(0,3,1,2)

print(tirage_prior.shape)

output = decoder_model(tirage_prior)

print(output.shape)

img = output[0]

img = img.permute(1, 2, 0)  # Réorganiser les dimensions (C, H, W) -> (H, W, C)
tab = img.detach().numpy()
npimg = (tab+1)/2  # Contraint toutes les valeurs dans l'intervalle [0, 1]
plt.imshow(npimg)  # Conversion en numpy pour matplotlib
# Affichage
plt.axis("off")  # Supprime les axes
plt.show()
```
